# Deterministic analysis
In order to run a deterministic analysis, a specific hazard need to be defined.

```python
import reXplan as rx
from datetime import date as dt_date

rx.config.path.SetWorkspaceFolder('Examples')
simulationName = 'case_single_hazard'
network = rx.network.Network(simulationName)
simulation = rx.simulation.Sim(simulationName)

network.event.hazardFromTrajectory('trajectory.csv', 
                                   max_intensity=60., max_radius=50., 
                                   sdate = dt_date(2022,4,1), edate = dt_date(2022,4,3), 
                                   geodata1 = rx.network.GeoData(47.4,5.8), geodata2 = rx.network.GeoData(54.9,15.0),
                                   delta_km=10, frequency='1H')
```

For the deterministic analysis simulation.initialize_model_sh() function should be used to initilize the simulation samples.

```python
simulation.initialize_model_sh(network, mc_iteration_factor = 50)
simulation.run(network, run_type = 'pm_ac_opf', delta = 1e-16, saveOutput = True)
```

The results of the OPF are stored in simulation.results as a DataFrame
